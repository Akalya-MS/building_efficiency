```python
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
import pandas as pd

# Load your dataset
data = pd.read_csv('Energy.csv')

# Separate features and target variable
X = data.drop(columns=['Heating_Load', 'Cooling_Load'])  # Features
y = data['Heating_Load']  # Target variable for classification, adjust as needed

# Encode target variable if it's not already encoded
# If the target variable is already encoded or in the desired format, you can skip this step
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Train LightGBM classifier
params = {
    'objective': 'multiclass',
    'num_class': len(label_encoder.classes_)  # Number of unique classes
}
train_data = lgb.Dataset(X_train, label=y_train)
model = lgb.train(params, train_data)

# Predict on the testing set
y_pred = model.predict(X_test)
y_pred = [int(x) for x in y_pred.argmax(axis=1)]  # Convert probabilities to class labels

# Create and print the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)


```
